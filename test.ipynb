```python
import psycopg2
from psycopg2.extras import RealDictCursor
import pandas as pd

conn = psycopg2.connect(
    dbname="instrumentsdb",
    user="postgres",
    password="1234",
    host="localhost",
    port=5432
)

with conn.cursor(cursor_factory=RealDictCursor) as cur:
    cur.execute("SELECT * FROM instruments ORDER BY name")
    rows = cur.fetchall()

conn.close()

df = pd.DataFrame(rows)
df
```

```python
import psycopg2
import pandas as pd

conn = psycopg2.connect(
    dbname="instrumentsdb",
    user="postgres",
    password="1234",
    host="localhost",
    port=5432
)

# Заявки
df_orders = pd.read_sql("SELECT * FROM best_offer_orders ORDER BY dt_place DESC", conn)
print("=== ЗАЯВКИ ===")
display(df_orders)

# Сделки
df_trades = pd.read_sql("SELECT * FROM best_offer_trades ORDER BY dt_trade DESC", conn)
print("=== СДЕЛКИ ===")
display(df_trades)

conn.close()
```

```python
import psycopg2

# параметры подключения — такие же как в instruments_db.py
conn = psycopg2.connect(
    host="localhost",
    port=5432,
    dbname="quik_robot",
    user="postgres",
    password="postgres"
)

isin = "RU000A0JX0J2"   # нужный инструмент
value = 15              # значение trades_curr

with conn:
    with conn.cursor() as cur:
        cur.execute(
            """
            UPDATE instruments
            SET trades_curr = %s
            WHERE isin = %s
            """,
            (value, isin)
        )

print(f"trades_curr для {isin} установлен в {value}")
```
